```python
import pandas as pd
import requests
```

```python
city = input("Enter the city")
apikey = 'add your API key'
url='https://api.openweathermap.org/data/2.5/forecast?q='+city+apikey
```

```python
r = requests.get(url)
json = r.json()
```

```python
json
```

```python
json.keys()
```

```python
pd.json_normalize(json)
```

```python
x = pd.json_normalize(json['list'])
```

```python
x
```

```python
df = x.filter(['main.temp','main.humidity','main.feels_like','main.temp_min','main.temp_max','main.pressure'],axis=1)
```

```python
df
```

```python
df.rename(columns={'main.temp':'temp'},inplace=True)
```

```python
df
```

```python
df.rename(columns={'main.humidity':'humidity'},inplace=True)
```

```python
df.rename(columns={'main.feels_like':'feels_like'},inplace=True)
```

```python
df.rename(columns={'main.min_temp':'min_temp'},inplace=True)
```

```python
df.rename(columns={'main.max_temp':'max_temp'},inplace=True)
```

```python
df.rename(columns={'main.pressure':'pressure'},inplace=True)
```

```python
df
```

```python
#Plotting Graphs for Analysis
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = df['temp']
y = df['humidity']
plt.scatter(x,y)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity")
plt.show()
```

```python
y1 = df['feels_like']
y2 = df['temp']
plt.plot(y1)
plt.plot(y2)
plt.title("Temperature vs Feels Like")
plt.show()
```

```python
y=df['pressure']
plt.plot(y)
plt.title('Pressure Graph')
plt.show()
```
